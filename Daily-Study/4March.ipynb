```python
# https://neetcode.io/problems/is-anagram/question?list=neetcode250
```

```python
class Solution:
    def isAnagram(self, s: str, t: str) -> bool:
        # using sorting , time complexity O(nlogn+mlogm) space complexity O(n) or O(1) depending on sorting algo
        if len(s) != len(t):
            return False
        return sorted(s)==sorted(t)
    
    def isAnagram1(self, s:str,t:str)-> bool :
        # using hash frequency map , time complexity O(n+m) space complexity  O(1) as it can not be more than 26 letters
        if len(s) != len(t):
            return False
        countS , countT = {},{}
        for i in range(len(s)):
            countS[s[i]] = 1+ countS.get(s[i],0)
            countT[t[i]]= 1+ countT.get(t[i],0)
        return countS == countT
    
    def isAnagram2(self, s: str, t: str) -> bool:
        # using sorting , time complexity O(nlogn+mlogm) space complexity O(n) or O(1) depending on sorting algo
        if len(s) != len(t):
            return False
        count =[0] *26
        print(f"display hash table : {count}")
        for i in range(len(s)):
            count[ord(s[i])-ord('a')] +=1
            count[ord(t[i])-ord('a')] -=1
        for val in count :
            if val != 0:
                return False
        return True

    
```

```python
obj = Solution()
print(f"Output using sorting : {obj.isAnagram('jar','raj')}")
print(f"Output using hash map frequency : {obj.isAnagram1('jar','raj')}")
print(f"Output using hash map frequency : {obj.isAnagram2('jam','raj')}")
```

```python
# Q2. https://neetcode.io/problems/two-integer-sum/question?list=neetcode250
```

```python
class Solution:
    def twoSum(self, nums: list[int], target: int) -> list[int]:
        # brute force approch time complexity will O(n*n) and space O(1)
        for i in range(len(nums)):
            for j in range(i+1,len(nums)):
                if nums[i]+nums[j]==target:
                    return [i,j]
        return []
    
    def twoSum1(self, nums: list[int], target: int) -> list[int]:
        # hash map approch time complexity will O(n) and space O(n)
        map = {}
        for i ,n  in enumerate(nums):
            map[n] = i
        print(f" display map : {map}")

        for i in range(len(nums)):
            if target-nums[i] in map and map[target-nums[i]] != i:
                return [i, map[target-nums[i]]]
        return []


    
```

```python
obj = Solution()
print(f"Output using brute force : {obj.twoSum([6,7,4],10)}")
print(f"Output using brute force : {obj.twoSum1([6,7,4],10)}")
```
