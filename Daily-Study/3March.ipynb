```python
# Q1  https://neetcode.io/problems/concatenation-of-array/question

class Solution:
    def getConcatenation(self, nums: list[int]) -> list[int]:
        ans = []
        for i in range(2):
            for num in nums:
                 ans.append(num)
        return ans
        
```

```python
obj = Solution()
ans = obj.getConcatenation([1,2,4])
print(ans)
```

```python
# Q2 . https://neetcode.io/problems/duplicate-integer/question

# Method 1 : Brute Force approch

class Solution:
    def hasDuplicate(self, nums: list[int])->bool:
        for i in range(len(nums)):
            for j in range(i+1,len(nums)):
                if nums[i]==nums[j]:
                    return True
        return False
    
    def hasDuplicate1(self, nums: list[int])->bool:
        # using sorting method which will give time complexilty O(nlogn)
        nums.sort()
        for i in range(1,len(nums)):
            if nums[i] ==nums[i-1]:
                return True
        return False 

    def hasDuplicate2(self, nums: list[int])->bool:
        # using hash set
        seen = set()
        for num in nums:
            if num in seen :
                return True
            seen.add(num)
        return False

    def hasDuplicate2(self, nums: list[int])->bool:
        # using hash set length as we know that set has no duplicate 
        if len(set(nums)) < len(nums):
            return True
        else :
            return False

    
       
```

```python
# Q . https://datalemur.com/questions/python-same-stripes

def is_same_stripes(matrix):
    diagonals = {}
    for i in range(len(matrix)):
        for j in range(len(matrix[0])):
            if (i-j) in diagonals and diagonals[i-j] != matrix[i][j]:
                return False
            diagonals[i-j]=matrix[i][j]
    return -1
   
```

```python
matrix = [[42, 7, 13, 99], [6, 42, 7, 13], [1, 6, 42, 7]]
print(is_same_stripes(matrix))
```
